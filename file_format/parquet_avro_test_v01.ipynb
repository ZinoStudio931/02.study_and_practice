# parguet, avro, csv comparative test 

```python
# make dataset for comparison
from sklearn.datasets import make_classification
import pandas as pd 

n_samples = int(1e+3)
num_X = 19

X, y = make_classification(
    n_samples=n_samples, 
    n_features=num_X
)

df = pd.DataFrame(X, columns=[f"col_{int(i+1)}" for i in range(num_X)])
df['y'] = y
```

### Reading and Writing Single Files

```python
import pyarrow as pa
import pyarrow.parquet as pq
```

```python
# Convert pd.DataFrame to parquet format
table = pa.Table.from_pandas(df)
table
```

```python
type(df), type(table)
```

```python
# Write pd.DataFrame to Parquet format 
pa.parquet.write_table(table, 'example.parquet')
# Write pd.DataFrame to csv format
df.to_csv('example.csv')
```

```python
# Read parquet file and convert to pd.DataFrame
table2 = pq.read_table('example.parquet')
table2.to_pandas()
```

```python
# Read parquet file with subset of columns 
pq.read_table('example.parquet', columns=['col_1', 'col_5', 'col_10', 'y'])
```

```python
pq.read_table('example.parquet', columns=['col_1', 'col_5', 'col_10', 'y']).to_pandas()
```
